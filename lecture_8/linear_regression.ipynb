````python
import numpy as np 

dob_array = np.array([
    1990, 1992, 1993, 1995, 1997, 1999, 2000, 
])

dob_array
````

````python
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
````

````python
# matplotlib scatter with trend

plt.plot(dob_array)
````

````python
# Create x values (indices)
x = np.arange(len(dob_array))
x
````

````python
# Create scatter plot
plt.scatter(x, dob_array, color='blue', label='Data points')

# Calculate trend line (linear regression)
slope, intercept = np.polyfit(x, dob_array, 1)
trend_line = slope * x + intercept

# Plot trend line
plt.plot(x, trend_line, color='red', label='Trend line')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Year of Birth')
plt.title('Scatter Plot with Trend Line')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
````

````python
import pandas as pd

df = pd.DataFrame({
    'index': np.arange(len(dob_array)),
    'Year': dob_array,
    'sex': np.random.choice(['M', 'F'], size=len(dob_array)),
})
df
````

````python
sns.regplot(
    data=df, 
    x='index', 
    y='Year', 
    scatter_kws={'color': 'blue'},  # Scatter point color
    line_kws={'color': 'red'}       # Trend line color
)

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Year of Birth')
plt.title('Scatter Plot with Trend Line (Seaborn)')
plt.grid(True)

# Show plot
plt.show()
````

````python
fig = px.scatter(
    df,
    x='index',
    y='Year',
    color='sex',
    trendline='ols',  # Ordinary Least Squares regression,
    title='Scatter Plot with Trend Line (Plotly)',
    labels={'Index': 'Position in Array', 'Year': 'Year of Birth'}
)

# Show interactive plot
fig.show()
````
